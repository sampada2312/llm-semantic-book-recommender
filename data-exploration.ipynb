```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False ,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman") #pearson correlation is used for continuous data, we have binary data of missing_description, thus spearman correlation method is used.

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
# checking to see how many rows have null values in description, num_pages, average_rating, published_year.
books[(books['description'].isna()) |
      (books['num_pages'].isna()) |
      (books['average_rating'].isna()) |
      (books['published_year'].isna())
]

# only 303 which is under 5% of the dataset, so we can drop those rows.
```

```python
books_missing = books[~(books['description'].isna()) &
      ~(books['num_pages'].isna()) &
      ~(books['average_rating'].isna()) &
      ~(books['published_year'].isna())
]
```

```python
book_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
# dropped rows/books which had description less than 25 works because the description wasn't meaningful and would have not given meaningful results.
```

```python
# combining title and subtitle cause subtitle is mostly empty but in a few places can add meaning
book_missing_25_words.loc[:, "title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(),
             book_missing_25_words["title"],
             book_missing_25_words["title"].astype(str) + ": " + book_missing_25_words["subtitle"].astype(str)))
```

```python
book_missing_25_words
```

```python
book_missing_25_words.loc[:, "tagged_description"] = (
    book_missing_25_words["isbn13"].fillna("").astype(str) + " " +
    book_missing_25_words["description"].fillna("").astype(str)
)
# adding isbn to description column
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)

# dropping all the columns that I created to understand the database.
```
